```python
# import required packages
import re
import pandas as pd
import numpy as np

# preprocess
from sklearn.feature_extraction.text import CountVectorizer

# local scripts
from text_utils import preprocess_corpus
```

```python
# load train set into dataframe
df = pd.read_csv('../../data/train_data.csv')

# printing the shape: (rows, columns)
display(df.shape)

# first 5 datapoints
display(df.head())
```

```python
# preprocess documents
# remove special characters, stopwords
# lemmatization
clean_headlines = preprocess_corpus(df.headline)

clean_headlines.head()
```

```python
# instantiate a CountVectorizer
cv = CountVectorizer()

# train and construct bag of words
counts = cv.fit_transform(clean_headlines)
```

```python
# extract bag of words as dataframe
df_bow = pd.DataFrame(counts.A, columns=cv.get_feature_names_out())

# include Sentiment column
df_bow['clickbait'] = df['clickbait']

# shape
display(df_bow.shape)

# first 5 datapoints
df_bow.head()
```
